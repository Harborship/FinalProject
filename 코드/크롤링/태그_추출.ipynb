```python
import pandas as pd
import numpy as np
import os
```

### csv 합치기

```python
final = pd.DataFrame(columns=['id'])
ran = int(len(os.listdir('시연크롤링/'))/3) +1
for i in range(1, ran):   
    content_df = pd.read_csv(f"시연크롤링/게시글{i}.csv")
    reply_df = pd.read_csv(f"시연크롤링/댓글{i}.csv")
    like_df = pd.read_csv(f"시연크롤링/좋아요{i}.csv")
    if i != 4:
        del content_df["Unnamed: 0"]
        del reply_df["Unnamed: 0"]
        del like_df["Unnamed: 0"]


    reply_lst = []
    num = len(reply_df.columns) if len(reply_df.columns) < len(reply_df.index) else len(reply_df.index)
    for i in reply_df.index[:num]:
        reply_lst.append(reply_df.iloc[i,i])
    reply_df_reshapd = pd.DataFrame( {"id" :reply_df.columns[:num], "reply":reply_lst}).dropna()

    like_lst = []
    num = len(like_df.columns) if len(like_df.columns) < len(like_df.index) else len(like_df.index)
    for i in like_df.index[:num]:
        like_lst.append(like_df.iloc[i,i])
    like_df_reshapd = pd.DataFrame( {"id" :like_df.columns[:num], "like":like_lst}).dropna()

    content_lst = []
    num = len(content_df.columns) if len(content_df.columns) < len(content_df.index) else len(content_df.index)
    for i in content_df.index[:num]:
        content_lst.append(content_df.iloc[i,i])
    content_df_reshapd = pd.DataFrame( {"id" :content_df.columns[:num], "content":content_lst}).dropna()
    
    final = pd.merge(final, pd.merge(pd.merge(content_df_reshapd,like_df_reshapd,how='outer', on='id'), reply_df_reshapd,how='outer', on='id'), how='outer')
```

```python
final =final.fillna("")
```

```python
len(final[["reply"]].values)
```

### 댓글 태그 추출

```python
reply_total_lst=[]
for reply in final[["reply"]].values.flatten():
    lst = []
    start = 0
    end = 0
    
    for i in range(len(reply)):
        if reply[i] == "#":
            start = i+1
            while(True):
                i+=1
                if i == len(reply):
                    lst.append(reply[start:])
                    break
                if reply[i] == " ":
                    lst.append(reply[start:i])
                    break
                if reply[i] == "#":
                    lst.append(reply[start:i])
                    break
    if len(lst) > 0:
        lst[-1] = lst[-1].replace("'","").replace("]","")
    reply_total_lst.append(lst)

```

```python
reply_total_lst
```

```python
content_total_lst=[]
for content in final[["content"]].values.flatten():
    lst = []
    start = 0
    end = 0
    for i in range(len(content)):
        if content[i] == "#":
            start = i+1
            while(True):
                i+=1
                if i == len(content):
                    lst.append(content[start:])
                    break
                if content[i] == " ":
                    lst.append(content[start:i])
                    break
                if content[i] == "#":
                    lst.append(content[start:i])
                    break
    if len(lst) > 0:
        lst[-1] = lst[-1].replace("'","").replace("]","")
    content_total_lst.append(lst)
```

```python
content_total_lst
```

### DataFrame으로 만들기

```python
tag_lst = pd.DataFrame({"id":final["id"].values, "reply_tag" : reply_total_lst, "content_tag" : content_total_lst})
```

```python
tot_tag = []
for i in tag_lst.index:
    tot_tag.append(tag_lst.iloc[i,1] + tag_lst.iloc[i,2])
len(tot_tag)
tag_lst_tot = pd.DataFrame({"id":final["id"].values, "tags" : tot_tag})

tag_lst_tot
```
